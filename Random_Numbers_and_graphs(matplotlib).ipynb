```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

plt.hist(np.random.randn(1000), bins=10, density=True, histtype = 'step', color='blue', label='Rand 01')
plt.hist(np.random.randn(10000), bins=100, density=True, histtype = 'step', color='red', label='Rand 02')
plt.legend(loc = 'upper left')
plt.axis([-5, 5, 0, 0.5])
plt.title('Histogram - Random Numbers')
plt.xlabel('Range')
plt.ylabel('Density')
plt.show()
```
